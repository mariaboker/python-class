```python
import pyautogui # automacao teclado e mouse
import time 
import pyperclip # copiar e colar via python

pyautogui.PAUSE = 1

#Passo 0
#Abrir navegador
#pyautogui.press("winleft")
#pyautogui.write("chrome")
#pyautogui.press("enter")
#pyautogui.alert("Vai começar, aperte OK e não mexa em nada")

#Passo 1
#Abrir uma nova aba
pyautogui.hotkey('ctrl', 't') #hotkey executa um atalho
#Entrar no link do sistema
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

#Passo 2
#Abrir pasta
time.sleep(5)
pyautogui.click(541, 399, clicks=2)
time.sleep(5)

#Passo 3
#Fazer download do arquivo
pyautogui.click(541,399) #apertar no arquivo
pyautogui.click(1612,226) #apertar em 3 pontinhos
pyautogui.click(1319,838) #apertar em fazer download
time.sleep(10) #tempo de download

#Passo 4
#ler o arquivo baixado
import pandas as pd

#Passo4
tabela = pd.read_excel(r"C:\Users\mmbok\Downloads\Vendas - Dez.xlsx") #r antes do caminho eh pra indicar que nao temos nenhum caractere espcial do python no nome
display(tabela) #print com melhor visualizacao

faturamento = tabela["Valor Final"].sum() #soma da columa valor final
quantidade = tabela["Quantidade"].sum() #soma da coluna quantidade
print(faturamento)
print(quantidade)

#Passo 5
#Enviar email

#Abrir uma nova aba
pyautogui.hotkey('ctrl', 't')
#Entrar no gmail
link = "https://mail.google.com/mail/u/0/#inbox"
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(5)

#Escrever email
pyautogui.click(52, 255) #apertar no +escrever
time.sleep(10)

email = "mmboker@gmail.com"
pyperclip.copy(email)
pyautogui.hotkey('ctrl', 'v')
time.sleep(3)
pyautogui.press("tab")
pyautogui.write("Email teste automacao") #assunto email
time.sleep(3)
pyautogui.press("tab")

#corpo do email - colocar entre 3 aspas para escrever em mais linhas
#obs. para usar caracteres especiais, usar copy
#se for usar variavel, colocar f antes das aspas e a variavel entre {} 
texto_email = f""" 
Este é um email teste de automacoes.

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs
Maria.
"""

pyautogui.write(texto_email)

#enviar email
pyautogui.hotkey("ctrl", "enter")
```
